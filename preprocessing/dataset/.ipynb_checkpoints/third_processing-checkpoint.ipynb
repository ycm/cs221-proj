```python
import pandas as pd
df = pd.read_csv("mand_kor_processed_latest.csv")
df = df.drop(df.index[62])
df.head()
```

```python
initials = ["ng", "gw", "kw", "b", "p", "m", "f", "d", "t", "l", "g", "k", "n", "h", "w", "z", "c", "s", "j"]
finals = ["aa", "aai", "aau", "aam", "aan", "aang", "aap", "aat", "aak", "ai", "au", "am", "an", \
          "ang", "ap", "at", "ak", "e", "ei", "eu", "em", "eng", "ep", "ek", "i", "iu", "im", "in", \
          "ing", "ip", "it", "ik", "o", "oi", "ou", "on", "ong", "ot", "ok", "u", "ui", "un", "ung", \
          "ut", "uk", "oe", "eoi", "eon", "oeng", "eot", "oek", "yu", "yun", "yut", "m", "ng"]
coda = ['p', 't', 'k', 'm', 'n', 'ng']
```

```python
def cantonizer(srt):
    if type(srt) == float:
        return ('','','','')
    tone = srt[-1]
    s = srt[:-1]
    l = ['','','', tone]
    for i in initials:
        if s.startswith(i):
            l[0] = i
            s = s[len(i):]
            break
    for i in finals:
        if s == i:
            l[1] = i
            break
    for i in coda:
        if l[1].endswith(i):
            l[1] = l[1][:-len(i)]
            l[2] = i
    if (l[0] == 'm' or l[0] == 'ng') and not l[1]:
        l[1] = l[0]
        l[0] = ''
#     print(l)
    assert(l[0] + l[1] + l[2] + l[3] == srt)
    assert(l[1] != '')
    assert(l[0] in initials or l[0] is '')
    assert(l[2] in coda or l[2] is '')
    return l
```

```python
li = df['Column4'].tolist()
```

```python
for i in range(len(li)):
    try:
        li[i] = cantonizer(li[i])
    except:
        print("Error: " + str(li[i]))
        li[i] = ('','','','')
```

```python
j1, j2, j3, j4 = zip(*li)
df = df.drop(df.columns[[5]], axis = 1)
```

```python
df.insert(loc=5, column='canto_tone', value=j4)
df.insert(loc=5, column='canto_coda', value=j3)
df.insert(loc=5, column='canto_nucl', value=j2)
df.insert(loc=5, column='canto_onset', value=j1)
df.head()
```

```python
initials = ['b', 'p', 'm', 'f', 'd', 't', 'n', 'l', 'g', 'k', 'h', 'j', 'q' \
           ,'x', 'zh', 'ch', 'sh', 'r', 'z', 'c', 's']
finals = ['i', 'e', 'a', 'ei', 'ai', 'ou', 'ao', 'en', 'an', 'ong', 'eng', 'ang' \
          ,'yi', 'ye', 'ya', 'ie', 'ia', 'you', 'iu', 'yao', 'iao', 'yin', 'in', 'yan', 'ian', \
          'yong', 'iong', 'ying', 'ing', 'yang', 'iang', 'wu', 'u', 'wo', 'uo', 'o', 'wa', 'ua', \
          'wei', 'ui', 'wai', 'uai', 'wen', 'un', 'wan', 'uan', 'weng', 'wang', 'uang',\
          'yu', 'ü', 'yue', 'üe', 'ue', 'yun', 'ün', 'üan', 'yuan']
coda = ['n', 'ng']
```

```python
def mandar(srt):
    if type(srt) == float:
        return ('','','')
    s = srt
    s = s.lower()
    if s == 'er':
        return ('', 'er', '')
    l = ['','','']
    for i in initials:
        if s.startswith(i):
            l[0] = i
            s = s[len(i):]
            break
    for i in finals:
        if s == i:
            l[1] = i
            break
    for i in coda:
        if l[1].endswith(i):
            l[1] = l[1][:-len(i)]
            l[2] = i
    assert(l[0] + l[1] + l[2] == srt.lower())
    assert(l[1] != '')
    assert(l[0] in initials or l[0] is '')
    assert(l[2] in coda or l[2] is '')
    return l
```

```python
li = df['mand_pron'].tolist()
```

```python
for i in range(len(li)):
    try:
        li[i] = mandar(li[i])
    except:
        print("Error: " + str(li[i]))
        li[i] = ('','','')
```

```python
j1, j2, j3 = zip(*li)
df = df.drop(df.columns[[4]], axis = 1)
```

```python
df.insert(loc=4, column='mando_coda', value=j3)
df.insert(loc=4, column='mando_nucl', value=j2)
df.insert(loc=4, column='mando_onset', value=j1)
df = df.drop(df.columns[[0]], axis = 1)
df.head()
```
