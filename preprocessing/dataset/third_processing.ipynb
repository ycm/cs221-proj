```python
import pandas as pd
df = pd.read_csv("mand_kor_processed_latest.csv")
df = df.drop(df.index[62])
df.head()
```

```python
initials = ["ng", "gw", "kw", "b", "p", "m", "f", "d", "t", "l", "g", "k", "n", "h", "w", "z", "c", "s", "j"]
finals = ["aa", "aai", "aau", "aam", "aan", "aang", "aap", "aat", "aak", "ai", "au", "am", "an", \
          "ang", "ap", "at", "ak", "e", "ei", "eu", "em", "eng", "ep", "ek", "i", "iu", "im", "in", \
          "ing", "ip", "it", "ik", "o", "oi", "ou", "on", "ong", "ot", "ok", "u", "ui", "un", "ung", \
          "ut", "uk", "oe", "eoi", "eon", "oeng", "eot", "oek", "yu", "yun", "yut", "m", "ng"]
coda = ['p', 't', 'k', 'm', 'n', 'ng']
```

```python
def cantonizer(srt):
    if type(srt) == float:
        return ('','','','')
    tone = srt[-1]
    s = srt[:-1]
    l = ['','','', tone]
    for i in initials:
        if s.startswith(i):
            l[0] = i
            s = s[len(i):]
            break
    for i in finals:
        if s == i:
            l[1] = i
            break
    for i in coda:
        if l[1].endswith(i):
            l[1] = l[1][:-len(i)]
            l[2] = i
    if (l[0] == 'm' or l[0] == 'ng') and not l[1]:
        l[1] = l[0]
        l[0] = ''
#     print(l)
    assert(l[0] + l[1] + l[2] + l[3] == srt)
    assert(l[1] != '')
    assert(l[0] in initials or l[0] is '')
    assert(l[2] in coda or l[2] is '')
    return l
```

```python
li = df['Column4'].tolist()
```

```python
for i in range(len(li)):
    try:
        li[i] = cantonizer(li[i])
    except:
        print("Error: " + str(li[i]))
        li[i] = ('','','','')
```

```python
def mandar(srt):
    if type(srt) == float:
        return ('','','')
    s = srt
    s = s.lower()
    if s == 'er':
        return ('', 'er', '')
    l = ['','','']
    for i in initials:
        if s.startswith(i):
            l[0] = i
            s = s[len(i):]
            break
    for i in finals:
        if s == i:
            l[1] = i
            break
    for i in coda:
        if l[1].endswith(i):
            l[1] = l[1][:-len(i)]
            l[2] = i
    assert(l[0] + l[1] + l[2] == srt.lower())
    assert(l[1] != '')
    assert(l[0] in initials or l[0] is '')
    assert(l[2] in coda or l[2] is '')
    return l
```

```python
def japaneet(srt):
    if type(srt) == float:
        return ('','','')
    s = srt
    l = ['','']
    for i in syllables:
        if s.startswith(i):
            l[0] = i
            s = s[len(i):]
            break
    l[1] = s
    assert(l[0] + l[1] == srt)
    assert(l[0] in syllables)
    return tuple(l)
```
